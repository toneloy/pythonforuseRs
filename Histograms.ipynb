# Histograms

**Problem:** we want to plot a histogram, change it's colors and transparency, add a title and axis labels.

 * **Solution in R:** we use `ggplot2`` 

 * **Solution in python:** we use `matplotlib.pyplot`

This one is actually not that different in both languages

```python
# Activate the interface between R and python and also load the rpy2 extension to use R in the notebook cells
from rpy2.robjects import r, pandas2ri
pandas2ri.activate()
%load_ext rpy2.ipython
```

## In R

```python
%%R

library(ggplot2)

ggplot(iris, aes(x = Sepal.Width)) + 
    geom_histogram(alpha = 0.5, fill = 'lightgrey', colour = 'grey', bins = 12) + 
    ggtitle('Sepal width histogram') + 
    xlab('Sepal width') + 
    ylab('Frequency') +
    theme_light()
```

## In python

```python
import matplotlib.pyplot as plt
%matplotlib inline

iris = r['iris']

plt.hist(iris['Sepal.Width'], 12, alpha=0.5, color='lightgrey', edgecolor='grey')
plt.title('Sepal width histogram')
plt.xlabel('Sepal width')
plt.ylabel('Frequency')
plt.show()
```

This one is pretty straight forward and the code very similar.
